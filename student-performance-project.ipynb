```python
import sqlite3
import pandas as pd

# Database connect
conn = sqlite3.connect("student.db")

# Cursor create
cursor = conn.cursor()

# Table create
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    student_id INTEGER,
    name TEXT,
    marks INTEGER
)
""")

# Data insert
cursor.execute("INSERT INTO students VALUES (1,'Khushi',90)")
cursor.execute("INSERT INTO students VALUES (2,'Rahul',75)")
cursor.execute("INSERT INTO students VALUES (3,'Anjali',88)")

conn.commit()

# SQL query run
query = "SELECT * FROM students"

df = pd.read_sql_query(query, conn)

df
```

```python
query = "SELECT * FROM students ORDER BY marks DESC LIMIT 1"

df = pd.read_sql_query(query, conn)

df
```

```python
query = "SELECT AVG(marks) AS Average_Marks FROM students"

df = pd.read_sql_query(query, conn)

df
```

```python
query = "SELECT MAX(marks) AS Highest_Marks FROM students"

df = pd.read_sql_query(query, conn)

df
```

```python
query = "SELECT * FROM students WHERE marks > 80"

df = pd.read_sql_query(query, conn)

df
```
